```python
import matplotlib.pyplot as plt

import numpy as np

x = np.array([1,2,3,4])

y = np.array(["chem","cse","maths","mech"])

plt.xlabel("Hours")

plt.ylabel("Subjects")

plt.plot(y,"o:r")

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np

x = np.array(["Python","Maths","English","chemistry"])

y = np.array([80,60,70,70])

myexplode = ([0.2,0,0,0])

mylabels = (["Python","Maths","English","chemistry"])

plt.pie(y,labels = mylabels,explode = myexplode)

plt.show()
```
